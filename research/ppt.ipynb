```python
capital_df = pd.DataFrame()
for nkey, pea in peas.items():
  capital_s = pd.Series(0.00, index=pg_keys + pc_keys)

  for key in pg_keys:
    p_unit = 'p'
    if key in p1_keys:
      p_unit = 'p1'
    if key in p0_keys:
      p_unit = 'p0'

    if key == 'c_pg_import':
      advance =  pea.get(ppv[key], type='import').advanceP().sum().sum() / 1e6
      revert = pea.get(ppv[key], type='export').revertP().sum().sum() / 1e6
      capital_s[key]  = revert + advance
    elif key == 'c_pg_inner_import':
      capital_s[key]  =  - pea.get(ppv[key], type='inner').innerExportP().sum().sum() / 1e6
    else:
      capital_s[key] = pea.get(ppv[key]).energy(p_unit) / 1e6

  for key in pc_keys:
    p_unit = 'p'
    if key in p1_keys:
      p_unit = 'p1'
    if key in p0_keys:
      p_unit = 'p0'

    if key == 'c_pc_export':
      advance =  - pea.get(ppv[key], type='export').advanceP().sum().sum() / 1e6
      revert = - pea.get(ppv[key], type='import').revertP().sum().sum() / 1e6
      capital_s[key]  = revert + advance
    elif key == 'c_pc_inner_export':
      capital_s[key]  =  - pea.get(ppv[key], type='inner').innerImportP().sum().sum() / 1e6
    else:
      capital_s[key] = - pea.get(ppv[key]).energy(p_unit) / 1e6


  capital_s = capital_s.rename(index= pp.map_name)
  capital_df[nkey] = capital_s

capital_df = capital_df.T
color_map = [pp.map_color[i] for i in pg_keys + pc_keys]


fig = px.bar(capital_df, color_discrete_map = pp.map_name_color, 
       title = "power balance in Germany", 
       )
fig.update_traces(
    texttemplate='%{y:.0f}',  # 显示整数，0位小数
    textposition='auto',  # 自动显示文本位置
)
fig.update_xaxes(title_text='Scenario')
fig.update_yaxes(title_text='power in TWh')

fig.update_layout(
    height=800 
)
```

```python
def plot_month_balance (pea, start = '2019-1-1 00:00:00', end = '2019-1-30 00:00:00', title= "01.2045"):
  flex_df = pd.DataFrame()
  for key in pg_keys:
    p_unit = 'p'
    if key in p1_keys:
      p_unit = 'p1'
    if key in p0_keys:
      p_unit = 'p0'

    if key == 'c_pg_import':
      advance =  pea.get(ppv[key], type='import').advanceP().sum(axis=1) / 1e3
      revert = pea.get(ppv[key], type='export').revertP().sum(axis=1) / 1e3
      flex_df[key]  = revert + advance
    elif key == 'c_pg_inner_import':
      flex_df[key]  =  pea.get(ppv[key], type='inner').innerExportP().sum(axis=1) / 1e3
    else:
      flex_df[key] = pea.get(ppv[key]).p(p_unit) / 1e3

  for key in pc_keys:
    p_unit = 'p'
    if key in p1_keys:
      p_unit = 'p1'
    if key in p0_keys:
      p_unit = 'p0'

    if key == 'c_pc_export':
      advance =  - pea.get(ppv[key], type='export').advanceP().sum(axis=1) / 1e3
      revert = - pea.get(ppv[key], type='import').revertP().sum(axis=1) / 1e3
      flex_df[key]  = revert + advance
    elif key == 'c_pc_inner_export':
      flex_df[key]  =  pea.get(ppv[key], type='inner').innerImportP().sum(axis=1) / 1e3
    else:
      flex_df[key] = - pea.get(ppv[key]).p(p_unit)  / 1e3

  flex_df = flex_df[start: end]
  flex_df.rename(columns=pp.map_name, inplace=True)

  fig = px.bar(flex_df, color_discrete_map = pp.map_name_color, 
          title = f"power balance in Germany {title}",
          )
  fig.update_yaxes(title_text='power in GW')
  fig.update_traces(marker_line_width=0)


  fig.update_layout(
      height=800,
      bargap=0 
  )
  return fig
```
